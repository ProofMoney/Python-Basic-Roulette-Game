```python
import random
import string
import os
import sys
 
past3,past2,past1 = 'N/A','N/A','N/A'
timesred,timesblack,timesgreen = 0,0,0
rednumbers = [1,3,5,7,9,12,14,16,18,19,21,23,25,27,30,32,34,36]
blacknumbers = [2,4,6,8,10,11,13,15,17,20,22,24,26,28,29,31,33,35]
green = [0]
validentries = ['r', 'R', 'b', 'B', 'g', 'G']
balance = 1000
bankruptcount = 0
losscount = 0
wincount = 1
while True:
    class general:
        def greb(astring, reqdpadding):
            w = len(str(astring))
            x = int(reqdpadding) - int(w)
            y = int(x) * ' '
            z = str(astring) + y
            return (z)
        def pass3(currentpast):
            global past3
            global past2
            global past1
            past3 = past2
            past2 = past1
            past1 = currentpast
            return (past3)
            return (past2)
            return (past1)
 
 
    balpadded = general.greb(balance, 15)
    losscountpadded = general.greb(bankruptcount, 7)
    timesredpad = general.greb(timesred, 9)
    timesblackpad = general.greb(timesblack, 9)
    timesgreenpad = general.greb(timesgreen, 9)
    past1pad = general.greb(past1, 16)
    past2pad = general.greb(past2, 16)
    past3pad = general.greb(past3, 16)
    def menutext():
        print("╔══════════════════════════╗\n"
              "║   Final Exam Roulette    ║\n"
              "║        MGM GRAND         ║\n"
              "║        Your Name         ║\n"
              "╠══════════════════════════╬══════════════════╗\n"
              "║ Available Inputs:        ║   Past Results   ║\n"
              "║ B/b = Blacks             ║ Red:   %s ║\n"
              "║ R/r = Reds               ║ Black: %s ║\n"
              "║ G/g = Green              ║ Green: %s ║\n"
              "╠══════════════════════════╬══════════════════╣" % (timesredpad, timesblackpad, timesgreenpad))
        print("║ Payouts:                 ║    Past Three    ║\n"
              "║ Blacks =  2x             ║ %s ║\n"
              "║ Reds   =  2x             ║ %s ║\n"
              "║ Greens =  14x            ║ %s ║" % (past1pad, past2pad, past3pad))
        print("╠══════════════════════════╬══════════════════╝\n"
              "║ Statistics:              ║\n"
              "║ "+"\033[1;32m"+"Balance: $%s" % (balpadded) +"\033[0m"+"║\n"
              "║ "+"\033[1;31m"+"Bankruptcy Count: %s" % (losscountpadded)+"\033[0m"+"║")
        print("╚══════════════════════════╝\n")
    if balance == 0:
        os.system("clear")
        menutext()
        print("You've gone bankrupt!  The MGM has given you $1000.")
        bankruptcount += 1
        balance += 1000
        irrelevant = input('Type Enter to continue.')
    else:
        os.system("clear")
        menutext()
        bet = input('Bet : ')
        if bet.isdigit():
            if int(bet) > balance:
                print('You dont have that much money!\n')
            else:
                balance -= int(bet)
                opt = input('\nColor: ')
                if opt in validentries:
                    rou = random.randrange(-0, 37)
                    if rou in rednumbers:
                        #past3 = past2
                        #past2 = past1
                        #past1 = 'RED'
                        general.pass3('RED')
                        timesred += 1
                        if opt == 'r':
                            balance += 2*int(bet)
                        elif opt == 'R':
                            balance += 2*int(bet)
                        else:
                            losscount += 1
                    elif rou in blacknumbers:
                        #past3 = past2
                        #past2 = past1
                        #past1 = 'BLACK'
                        general.pass3('BLACK')
                        timesblack += 1
                        if opt == 'b':
                            balance += 2*int(bet)
                        elif opt == 'B':
                            balance += 2*int(bet)
                        else:
                            losscount += 1
                    elif rou in green:
                        #past3 = past2
                        #past2 = past1
                        #past1 = 'GREEN'
                        general.pass3('GREEN')
                        timesgreen += 1
                        if opt == 'g':
                            balance += 14*int(bet)
                        elif opt == 'G':
                            balance += 14*int(bet)
                        else:
                            losscount += 1
                else:
                    print('Invalid Entry')
                    balance += int(bet)
       
        else:
            print('Invalid Entry')
```
